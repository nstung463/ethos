```python
import re

def clean_user_context(text):
    # Định nghĩa các section quan trọng cần giữ lại
    important_sections = [
        "Demographics Information",
        "Interests & Preferences",
        "Dated Events, Projects & Plans",
        "User Correction Ledger"
    ]
    
    cleaned_data = {}
    
    # 1. Trích xuất các đề mục lớn và nội dung đi kèm
    for section in important_sections:
        # Regex tìm kiếm từ tiêu đề section cho đến khi gặp tiêu đề tiếp theo hoặc hết văn bản
        pattern = rf"(\*\*{section}\*\*|# {section})(.*?)(?=(?:\n# |\n\*\*|$))"
        match = re.search(pattern, text, re.DOTALL)
        
        if match:
            content = match.group(2).strip()
            # 2. Loại bỏ các dòng "Evidence" (vì LLM chỉ cần kết luận, không cần bằng chứng hội thoại)
            content = re.sub(r"^\s*\* Evidence:.*$", "", content, flags=re.MULTILINE)
            # 3. Loại bỏ các mốc thời gian "Conversation Date"
            content = re.sub(r"Conversation Date: \d{4}-\d{2}.*$", "", content, flags=re.MULTILINE)
            # 4. Làm sạch khoảng trắng thừa
            content = "\n".join([line.strip() for line in content.split('\n') if line.strip()])
            
            cleaned_data[section] = content

    return cleaned_data

# Giả sử 'raw_text' là biến chứa nội dung User Summary của bạn
result = clean_user_context(raw_text)

Cách in ra để đưa vào Prompt cho LLM khác:
for key, value in result.items():
    print(f"### {key}\n{value}\n")
```
